```python
import pandas as pd
import numpy as np

from sklearn.impute import KNNImputer
```

```python
df = pd.read_csv('life_expectancy.csv')
df.head()
```

```python
#dropping uncessary columns
#inplace le chai value replace gardinxa
df.drop(columns=['id'], inplace=True)
df.columns
```

```python
df.head()
```

```python
#handeling duplicates
#duplicates vaneko dueta rows identical vayo vane

print(df.duplicated().sum())
df.drop_duplicates(inplace=True)
print(df.drop_duplicated().sum())
```

```python
df.isnull().sum()
```

```python
#dropping rows
df.dropna(subset=['life_expectancy'], inplace=True)
df.isnull().sum()
```

```python
#filling

df['adult_mortality'] = df['adult_mortality'].fillna(0)
print(df.isnull().sum())
```

```python
df['country'] = df['country'].fillna(df['country'].mode()[0])
print(df.isnull().sum())
```

```python
df['bmi'] = df['bmi'].fillna(df['bmi'].mean())
print(df.isnull().sum())
```

```python
df.head()
```

```python
#filling average
cols_for_mean = ['total_expenditure', 'gdp']
for cols in cols_for_mean:
    df[cols] = df[cols].fillna(df[cols].mean())
df.isnull().sum()
```

```python
#imputing using KNN Imputer
#KNNImputer le chai afno najik ko value haldinxa

cols_for_impute = ['population', 'schooling']
ki = KNNImputer()
for cols in cols_for_impute:
    df[cols] = ki.fit_transform(df[[cols]])
df.isnull().sum()
```

```python
df
```

```python
#level encoding
pd.factorize(df['country'])
df['country'] = pd.factorize(df['country'])[0]
df.head()
```

```python
#one hot encoding

df['status'].unique()
```

```python
df = pd.get_dummies(data=df, columns=['status'], drop_first=True)
df.head()
```
